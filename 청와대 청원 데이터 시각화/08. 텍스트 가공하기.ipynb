# 8. 텍스트 가공하기

```python
from konlpy.tag import Kkma
import collections
kkma = Kkma()
```

```python
list_sentence = [
    "저는 여자친구가 있어요",
    "저는 남자친구가 있어요",
    "아버지가 방에 들어가신다.",
    "아버지 가방에 들어가신다.",
    "돈 많은 백수가 되고 싶다.",
    "일더하기 일은 귀요미",
    "이더하기 이는 귀요미",
    "삼더하기 삼은 귀요미",
    "사더하기 일사는 귀요미",
    "오더하기 오는 귀요미"
]
```

```python
for i in list_sentence:
    print(i)
```

```python
list_result = []

for i in list_sentence :
    print(kkma.nouns(i))
    list_result = list_result+kkma.nouns(i)
    print("------------------------------------------------")
    print("list_result =", list_result)
```

```python
print(list_result)
```

```python
print(collections.Counter(list_result))
```

```python
collections.Counter(list_result).most_common(5)
```
